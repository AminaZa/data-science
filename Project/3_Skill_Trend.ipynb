```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset= load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x)else x)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst')&(df['job_country']=='United States')].copy()
df_DA_US['job_posted_month_nb']=df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_nb' ,columns='job_skills' , aggfunc='size', fill_value=0)
df_DA_US_pivot.loc['total']=df_DA_US_pivot.sum()
df_DA_US_pivot=df_DA_US_pivot[df_DA_US_pivot.loc['total'].sort_values(ascending=False).index]
df_DA_US_pivot.drop('total', inplace=True)
```

```python
df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_nb').size()
```

```python
df_DA_US_perc = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
df_DA_US_perc = df_DA_US_perc.reset_index()
df_DA_US_perc['job_posted_month'] = df_DA_US_perc['job_posted_month_nb'].apply(lambda x: pd.to_datetime(x,format='%m').strftime('%b'))
df_DA_US_perc=df_DA_US_perc.set_index('job_posted_month')
df_DA_US_perc.drop(columns='job_posted_month_nb',inplace=True)

df_DA_US_perc
```

```python
import numpy as np
df_plot = df_DA_US_perc.iloc[:,:5]
sns.lineplot(data = df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')

plt.title('Trending Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2025')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax=plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
# Get last x and y values
last_x = 11.2
y_positions = df_plot.iloc[-1].values
labels = df_plot.columns

# Sort by y so we place labels from bottom to top
sorted_indices = np.argsort(y_positions)

# Minimum vertical spacing between labels
min_gap = 1.5  # adjust based on your y-scale

adjusted_y = y_positions.copy()
for idx in sorted_indices[1:]:
    # Check if this label is too close to the previous one
    if adjusted_y[idx] - adjusted_y[sorted_indices[0]] < min_gap:
        adjusted_y[idx] = adjusted_y[sorted_indices[0]] + min_gap
    sorted_indices[0] = idx  # update reference

# Plot labels with adjusted positions
for i, label in enumerate(labels):
    plt.text(last_x, adjusted_y[i], label)
```
